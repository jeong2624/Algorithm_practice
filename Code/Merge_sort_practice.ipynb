### Merge two sorted sublists to one sorted list

Input : [1, 5, 8], [2, 3, 6]

Output : [1, 2, 3, 5, 6, 8]

```python
def merge(A : list, B : list):
    output = []
    i = 0; j = 0

    while (i < len(A)) and (j < len(B)):
        if A[i] < B[j]:
            output.append(A[i])
            i += 1
        elif A[i] > B[j]:
            output.append(B[j])
            j += 1

    if i < j:
        output.append(A[i])
    else:
        output.append(B[j])

    print(output)
```

```python
A = [1, 5, 8]
B = [2, 3, 6]
```

```python
merge(A, B)
```

### Implement merge sort

```python
A = [54, 23, 79, 13]
```

```python
def MergeSort(A : list):
    if len(A) <= 1:
        return A

    mid = len(A) // 2

    left = A[:mid]
    right = A[mid:]

    leftlist = MergeSort(left)
    rightlist = MergeSort(right)

    return Merge(leftlist, rightlist)
```
