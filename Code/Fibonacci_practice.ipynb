### Dynamic programming

```python
# Fibonacci sequence by suing recursive algorithms

def fib1(n: int):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib1(n-1) + fib1(n-2)
```

```python
n = int(input("Input integer : "))

print(fib1(n))
```

```python
# Fibonacci sequence by using dynamic programming

def fib2(n: int):
    fib = [0 * i for i in range(n+1)]

    fib[0] = 0
    fib[1] = 1

    for i in range(2, n+1):
        fib[i] = fib[i-1] + fib[i-2]

    return fib[n]
```

```python
n = int(input("Input integer : "))

print(fib2(n))
```
