### Implement a python function for swapping two variables

(ex) x = 3, y = 8 --> swap(x, y) --> x = 8, y = 3

```python
# Load python tutor
from metakernel import register_ipython_magics
register_ipython_magics()
```

```python
%%tutor

def swap(x : int, y : int):
    tmp = x
    x = y
    y = tmp
    print("x =", x, ",", "y =", y)

swap(x = 3, y = 8)
```

### Bubble sort

![nn](https://gmlwjd9405.github.io/images/algorithm-bubble-sort/bubble-sort.png)

```python
%%tutor

def Bubble_sort(A : list):
    n = len(A)

    for i in range(n-1):
        for j in range(0, n-i-1):
            if A[j] > A[j+1]:
                A[j], A[j+1] = A[j+1], A[j]
    print(A)

A = [5, 2, 4, 1, 3]
Bubble_sort(A)
```
