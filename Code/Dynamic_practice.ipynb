# Dynamic Programming

### Binominal Coefficient (이항계수)

```python
def comb(n : int, k : int):
    C = [[0 for j in range(k + 1)] for i in range(n + 1)]

    for i in range(1, n + 1):
        for j in range(min(i, k) + 1):
            if j == 0 or j == i:
                C[i][j] = 1
            else:
                C[i][j] = C[i-1][j-1] + C[i-1][j]

    return C[n][k]
```

```python
print(comb(5, 3))
```

```python
print(comb(5, 5))
```

```python
print(comb(5, 0))
```

### Matrix chain Multiplication (행렬의 곱셈순서)

```python
import sys

def MatrixChainOrder(n : int, d : list):

    M = [[0 * x for x in range(n)] for x in range(n)]

    for diag in range(1, n):
        for i in range(1, n - diag):
            j = i + diag
            M[i][j] = sys.maxsize # Initialize inf.
            for k in range(i, j):
                M[i][j] = min(M[i][j],
                              M[i][k] + M[k + 1][j] + d[i-1] * d[k] * d[j])

    return M[1][n-1]

# A : 10 * 30 matrix, B : 30 * 5 matrix, C : 5 * 60 matrix

d = [10, 30, 5, 60] # d0 = 10, d1 = 30, d2 = 5, d3 = 60
n = 4
```

```python
print(MatrixChainOrder(n, d)) # (AB)C = (10 * 30 * 5) + (10 * 5 * 60) = 1500 + 3000 = 4500
```
