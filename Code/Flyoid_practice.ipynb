### Flyoid's algorithms

```python
INF = sys.maxsize

def Floyd(n : int, A : list):

    D = [[INF] * (n) for _ in range(n)]

    for i in range(n):
        for j in range(n):
            if i == j:
                D[i][j] = 0
            else:
                D[i][j] = A[i][j]


    for k in range(n):
        for i in range(n):
            for j in range(n):
                if D[i][j] > D[i][k] + D[k][j]:
                    D[i][j] = D[i][k] + D[k][j]

    return D
```

```python
n = 4
G = [[0, 6, 10, 2],
     [INF, 0, INF, 5],
     [3, 12, 0, INF],
     [INF, INF, 7, 0]]

print(Floyd(n, G))
```
