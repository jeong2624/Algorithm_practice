### Implement linear search

```python
def linear_search(A : list, x : int):
    n = len(A)

    for i in range(n):
        if A[i] == x:
            return i

    return None
```

```python
A = [15, 45, 31, 16, 27, 12, 8]
```

```python
print(linear_search(A, 27))
```

```python
print(linear_search(A, 3))
```

### Implement binary search

```python
def binary_search(A : list, low : int, high : int, x : int):
    mid = (low + high) // 2

    if high > low:
        if A[mid] == x:
            return mid
        elif A[mid] < x:
            binary_search(A, low, mid, x)
        else:
            binary_search(A, mid + 1, high, x)
    return None
```

```python
print(binary_search(A, 1, len(A), 27))
```

```python
print(binary_search(A, 1, len(A), 3))
```
