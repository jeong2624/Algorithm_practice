#### 후위표기법 도전과제 1

조건 :
1. 한 자리의 양의 정수만 입력으로 들어온다고 가정.


2. +, -, *, /의 사칙연산만 고려하고 괄호는 고려하지 않음.


3. Examples
    * Examples 1 : 5 + 6 * 3 --> 563*+

    * Examples 2 : 5 + 6 - 3 --> 563+-

```python
def no_bracket_postfix(equation: str):
    text = list(equation)
    stack = []
    result = []

    for t in text:
        if t.isdigit():
            result.append(t)
        else:
            stack.append(t)

    if stack[-1] == "*" or stack[-1] == "/":
        result.extend(stack.pop())
        result.extend(stack)
    else:
        result.extend(stack)

    return result
```

```python
equation = input("Input the equation : ")
print(no_bracket_postfix(equation))
```

#### 후위표기법 도전과제 2

조건 :
1. 한 자리의 양의 정수만 입력으로 들어온다고 가정.


2. +, -, *, /의 사칙연산과 괄호가 입력으로 들어올 수 있음.


3. Examples
    * Examples 1 : (5 + 6) * 3 --> 56+3*

    * Examples 2 : (3 + 5) * (7 + 2) --> 35+72+*

```python
def bracket_postfix(equation: str):
    text = list(equation)
    stack = []
    result = []

    for t in text:
        if t.isdigit():
            result.append(t)
            while stack and (stack[-1] != '(' and stack[-1] != ')'):
                result.extend(stack.pop())
        else:
            if t == "(":
                stack.append(t)
            elif t == "+" or t == "-":
                while stack and stack[-1] == "(":
                    stack.pop()
                stack.append(t)
            elif t == "*" or t == "/":
                while stack and (stack[-1] == "+" or stack[-1] == "-"):
                    stack.pop()
                stack.append(t)
            elif t == ")":
                while stack and (stack[-1] == '+' or stack[-1] == '-'):
                    stack.pop()

    return "".join(result)
```

```python
equation = input("Input the equation : ")
print(bracket_postfix(equation))
```
